```python
# training_image_visualization.py
import os
import matplotlib.pyplot as plt
from PIL import Image

# Define your dataset directory
dataset_dir = r"D:\plastiphalt_ai_project\dataset"  # raw string avoids backslash escape issues
classes = ["PET", "HDPE", "LDPA", "PP", "PS", "PVC", "OTHER"]

# Grid settings
cols = 3
rows = (len(classes) + cols - 1) // cols  # calculate rows needed
plt.figure(figsize=(12, 8))

for i, cls in enumerate(classes):
    img_dir = os.path.join(dataset_dir, cls)
    img_files = [f for f in os.listdir(img_dir) if f.lower().endswith(('.png', '.jpg', '.jpeg'))]
    
    if not img_files:
        print(f"No images found for class: {cls}")
        continue

    img_path = os.path.join(img_dir, img_files[0])
    img = Image.open(img_path)

    plt.subplot(rows, cols, i + 1)
    plt.imshow(img)
    plt.title(cls, fontsize=10)
    plt.axis('off')

plt.suptitle("Sample Training Images of Plastic Categories", fontsize=16)
plt.tight_layout()
plt.subplots_adjust(top=0.92)

# Save for LaTeX
plt.savefig("training_samples.png", dpi=300, bbox_inches='tight')
plt.show()
```
